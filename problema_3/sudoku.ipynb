# Descripci√≥n
```
El programa selecciona un tablero de Sudoku entre cinco opciones disponibles. El usuario ingresa valores en coordenadas seleccionadas. Al final, se verifica si el tablero fue completado correctamente.
```

# Pseudoc√≥digo
```
Inicio
    Limpiar pantalla
    Mostrar t√≠tulo del juego

    Definir un tablero vac√≠o de 9x9
    Llenar el tablero usando backtracking

    Mostrar opciones de dificultad
    Leer dificultad seleccionada
    Eliminar n√∫meros del tablero seg√∫n la dificultad

    Mientras el tablero no est√© completo hacer
        Limpiar pantalla
        Mostrar reglas del Sudoku
        Mostrar t√≠tulo
        Imprimir el tablero

        Pedir al usuario la fila donde quiere ingresar un n√∫mero
        Pedir al usuario la columna donde quiere ingresar un n√∫mero

        Si la celda ya tiene un n√∫mero entonces
            Mostrar mensaje de error
            Continuar con la siguiente iteraci√≥n

        Pedir al usuario el n√∫mero a colocar en la celda

        Si el n√∫mero es v√°lido seg√∫n las reglas del Sudoku entonces
            Colocar el n√∫mero en la celda
        Sino
            Mostrar mensaje de error

    Limpiar pantalla
    Mostrar t√≠tulo
    Imprimir tablero final
    Mostrar mensaje de √©xito
Fin
```

````python
import os  # Importa la librer√≠a os para interactuar con el sistema operativo
import random  # Importa la librer√≠a random para generar n√∫meros aleatorios

def limpiar_pantalla():
    """Limpia la pantalla en diferentes sistemas operativos."""
    os.system("cls" if os.name == "nt" else "clear")  # Limpia la pantalla seg√∫n el sistema operativo

def mostrar_titulo():
    """Muestra el t√≠tulo del juego."""
    print("=" * 75)  # Imprime una l√≠nea de separaci√≥n
    print(" " * 30, "JUEGO DE SUDOKU")  # Imprime el t√≠tulo del juego centrado
    print("=" * 75)  # Imprime otra l√≠nea de separaci√≥n

def mostrar_reglas():
    # Funci√≥n para mostrar las reglas del juego Sudoku en la consola
    print("=" * 75)  # Imprime una l√≠nea de separaci√≥n
    print(" " * 28, "REGLAS DEL SUDOKU")  # Imprime un encabezado de las reglas con salto de l√≠nea inicial
    print("=" * 75)  # Imprime otra l√≠nea de separaci√≥n
    print("1. Se genera un tablero de Sudoku 9x9 con algunos espacios vac√≠os.")  # Explica la primera regla
    print("2. Debes rellenar los espacios vac√≠os con n√∫meros del 1 al 9.")  # Explica la segunda regla
    print("3. No puedes repetir n√∫meros en la misma fila, columna o subcuadr√≠cula 3x3.")  # Explica la tercera regla
    print("4. Ganas cuando completas el tablero correctamente.")  # Explica la cuarta regla
    print("5. Guia de ingreso de datos C para columnas y F para Filas")  # Explica la quinta regla
    print("=" * 75)  # Imprime una l√≠nea divisoria para finalizar la secci√≥n de reglas

def imprimir_tablero(tablero):
    """Imprime el tablero de Sudoku con formato estructurado."""
    print("\n  C 0 1 2 | 3 4 5 | 6 7 8")  # Imprime el encabezado de las columnas
    print("F ‚îå-----------------------‚îê")  # Imprime la parte superior del tablero

    for fila in range(9):  # Itera sobre cada fila del tablero
        if fila in [3, 6]:  # Imprime un separador horizontal cada 3 filas
            print("  |-------|-------|-------|")  # Separador de bloques

        print(f"{fila} |", end=" ")  # Imprime el n√∫mero de la fila y el borde izquierdo del tablero
        for col in range(9):  # Itera sobre cada columna de la fila
            if col in [3, 6]:  # Imprime un separador vertical cada 3 columnas
                print("|", end=" ")  # Separador vertical de bloques

            print(tablero[fila][col] if tablero[fila][col] != 0 else ".", end=" ")  # Imprime el n√∫mero o un punto si la celda est√° vac√≠a

        print("|")  # Imprime el borde derecho del tablero

    print("  ‚îî-----------------------‚îò")  # Imprime la parte inferior del tablero

def generar_subtablero(tablero, fila, col, num):
    """Verifica si un n√∫mero puede colocarse en la posici√≥n dada."""
    if num in tablero[fila]:  # Verifica si el n√∫mero ya est√° en la fila
        return False

    for i in range(9):  # Verifica si el n√∫mero ya est√° en la columna
        if tablero[i][col] == num:
            return False

    inicio_fila, inicio_col = (fila // 3) * 3, (col // 3) * 3  # Calcula el inicio de la subcuadr√≠cula 3x3
    for i in range(3):  # Itera sobre las filas de la subcuadr√≠cula
        for j in range(3):  # Itera sobre las columnas de la subcuadr√≠cula
            if tablero[inicio_fila + i][inicio_col + j] == num:  # Verifica si el n√∫mero ya est√° en la subcuadr√≠cula
                return False

    return True  # Retorna True si el n√∫mero puede colocarse en la posici√≥n dada

def generar_tablero(tablero):
    """Verifica si el tablero est√° lleno."""
    for fila in tablero:  # Itera sobre cada fila del tablero
        if 0 in fila:  # Verifica si hay alguna celda vac√≠a (con valor 0)
            return False
    return True  # Retorna True si el tablero est√° lleno

def generar_sudoku():
    """Genera un Sudoku completo v√°lido."""
    def resolver_sudoku(tablero):
        """Resuelve el Sudoku con backtracking."""
        for fila in range(9):  # Itera sobre cada fila del tablero
            for col in range(9):  # Itera sobre cada columna de la fila
                if tablero[fila][col] == 0:  # Verifica si la celda est√° vac√≠a
                    for num in range(1, 10):  # Intenta colocar n√∫meros del 1 al 9
                        if generar_subtablero(tablero, fila, col, num):  # Verifica si el n√∫mero puede colocarse
                            tablero[fila][col] = num  # Coloca el n√∫mero en la celda
                            if resolver_sudoku(tablero):  # Llama recursivamente para resolver el resto del tablero
                                return True
                            tablero[fila][col] = 0  # Si no se puede resolver, vac√≠a la celda y prueba con otro n√∫mero
                    return False  # Retorna False si no se puede colocar ning√∫n n√∫mero en la celda
        return True  # Retorna True si el tablero est√° resuelto

    tablero = [[0 for _ in range(9)] for _ in range(9)]  # Inicializa un tablero vac√≠o (9x9) con ceros
    resolver_sudoku(tablero)  # Llama a la funci√≥n para resolver el Sudoku
    return tablero  # Retorna el tablero resuelto

def eliminar_numeros(tablero, porcentaje):
    """Elimina n√∫meros del tablero seg√∫n la dificultad."""
    num_casillas_a_vaciar = int(81 * porcentaje)  # Calcula el n√∫mero de celdas a vaciar seg√∫n el porcentaje
    casillas = [(fila, col) for fila in range(9) for col in range(9)]  # Crea una lista con todas las posiciones del tablero
    random.shuffle(casillas)  # Mezcla aleatoriamente las posiciones

    for _ in range(num_casillas_a_vaciar):  # Itera sobre el n√∫mero de celdas a vaciar
        fila, col = casillas.pop()  # Toma una posici√≥n aleatoria de la lista
        tablero[fila][col] = 0  # Vac√≠a la celda en la posici√≥n seleccionada

def generar_dificultad():
    """Permite seleccionar la dificultad."""
    niveles = {
        "1": ("S√∫per f√°cil", 3/81),  # Define el nivel "S√∫per f√°cil" con su porcentaje de celdas vac√≠as
        "2": ("F√°cil", 0.30),  # Define el nivel "F√°cil" con su porcentaje de celdas vac√≠as
        "3": ("Medio", 0.45),  # Define el nivel "Medio" con su porcentaje de celdas vac√≠as
        "4": ("Dif√≠cil", 0.70)  # Define el nivel "Dif√≠cil" con su porcentaje de celdas vac√≠as
    }

    print("\nSelecciona la dificultad:")  # Imprime el mensaje para seleccionar la dificultad
    for clave, (nombre, _) in niveles.items():  # Itera sobre los niveles de dificultad
        print(f"{clave}. {nombre}")  # Imprime cada nivel de dificultad con su clave

    while True:  # Bucle infinito para solicitar una opci√≥n v√°lida
        opcion = input("Opci√≥n: ")  # Solicita la opci√≥n al usuario
        if opcion in niveles:  # Verifica si la opci√≥n es v√°lida
            return niveles[opcion][1]  # Retorna el porcentaje correspondiente al nivel seleccionado
        print("Opci√≥n no v√°lida, intenta nuevamente.")  # Imprime un mensaje de error si la opci√≥n no es v√°lida

def obtener_dificultad(mensaje, min_val, max_val):
    """Solicita un n√∫mero al usuario y maneja errores."""
    while True:  # Bucle infinito para solicitar un n√∫mero v√°lido
        try:
            num = int(input(mensaje))  # Solicita el n√∫mero al usuario y lo convierte a entero
            if min_val <= num <= max_val:  # Verifica si el n√∫mero est√° dentro del rango permitido
                return num  # Retorna el n√∫mero si es v√°lido
            print(f"Error: Ingresa un n√∫mero entre {min_val} y {max_val}.")  # Imprime un mensaje de error si el n√∫mero est√° fuera del rango
        except ValueError:
            print("Error: Debes ingresar un n√∫mero v√°lido.")  # Imprime un mensaje de error si la entrada no es un n√∫mero v√°lido

def jugar_sudoku():
    """Funci√≥n principal para jugar Sudoku."""
    limpiar_pantalla()  # Limpia la pantalla
    mostrar_titulo()  # Muestra el t√≠tulo del juego

    tablero = generar_sudoku()  # Genera un tablero de Sudoku completo
    porcentaje_vacio = generar_dificultad()  # Solicita al usuario seleccionar la dificultad
    eliminar_numeros(tablero, porcentaje_vacio)  # Elimina n√∫meros del tablero seg√∫n la dificultad seleccionada

    while not generar_tablero(tablero):  # Bucle que se ejecuta mientras el tablero no est√© lleno
        limpiar_pantalla()  # Limpia la pantalla
        mostrar_reglas()  # Muestra las reglas del juego
        mostrar_titulo()  # Muestra el t√≠tulo del juego
        imprimir_tablero(tablero)  # Imprime el tablero actual

        print("\nIngrese las coordenadas y el n√∫mero a colocar:")  # Solicita al usuario ingresar las coordenadas y el n√∫mero

        fila = obtener_dificultad("Ingresa la fila (0-8): ", 0, 8)  # Solicita la fila al usuario
        col = obtener_dificultad("Ingresa la columna (0-8): ", 0, 8)  # Solicita la columna al usuario

        if tablero[fila][col] != 0:  # Verifica si la celda ya tiene un n√∫mero
            print("Error: La celda ya tiene un n√∫mero. Intenta en otra posici√≥n.")  # Imprime un mensaje de error
            continue  # Contin√∫a con la siguiente iteraci√≥n del bucle

        num = obtener_dificultad("Ingresa un n√∫mero (1-9): ", 1, 9)  # Solicita el n√∫mero al usuario

        if generar_subtablero(tablero, fila, col, num):  # Verifica si el n√∫mero puede colocarse en la posici√≥n dada
            tablero[fila][col] = num  # Coloca el n√∫mero en la celda
        else:
            print("Error: N√∫mero no v√°lido en esta posici√≥n. Intenta de nuevo.")  # Imprime un mensaje de error

    limpiar_pantalla()  # Limpia la pantalla
    mostrar_titulo()  # Muestra el t√≠tulo del juego
    imprimir_tablero(tablero)  # Imprime el tablero completo
    print("\n¬°Felicidades! Has completado el Sudoku correctamente.")  # Imprime un mensaje de felicitaci√≥n

def main():
    jugar_sudoku()  # Llama a la funci√≥n principal para jugar Sudoku

if __name__ == "__main__":
    main()  # Ejecuta la funci√≥n main si el script se ejecuta directamente
````
